#### 03 - Remodele os dados para trabalhar com trimestres e ano, e plote um gráficos de vendas por trimestre e ano

```python
resample_3M = series.resample('3M')

vendas_medias_tres_meses = resample_3M.mean()
print(vendas_medias_tres_meses.head())
```

```python
from matplotlib import pyplot
vendas_medias_tres_meses.plot(figsize=(10,5), linewidth=3, fontsize=10, title='Vendas médias a cada três meses')
pyplot.show()
```
