```python
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score


previsores = pd.read_csv('../input/entradas-breast/entradas_breast.csv')
classe = pd.read_csv('../input/saidas-breast/saidas_breast.csv')

def criarRede():
    classificador = Sequential()

    classificador.add(Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform', input_dim = 30))
    
    # Dropout ---> tcnica utilizada para evitar overfitting no treinamento da base, 
    # zerando um percentual de neurnios da camada de entrada (ou oculta)
    # Os valores zerados no tero influncia sobre o resultado final
    # Se utilizar um valor muito alto no dropout, ocorrer underfitting (a rede neural no vai conseguir aprender)
    # Dropout = 0.2 ---> zera o valor de 20% dos 30 neurnios da camada de entrada (zera o valor de 6 neurnios da camada de entrada)
    classificador.add(Dropout(0.2))
    
    classificador.add(Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform'))
    
    # Dropout = 0.2 ---> zera o valor de 20% dos 16 neurnios da primeira camada oculta
    classificador.add(Dropout(0.2))
    
    classificador.add(Dense(units = 1, activation = 'sigmoid'))
    
    otimizador = keras.optimizers.Adam(lr = 0.001, decay = 0.0001, clipvalue = 0.5)
    
    classificador.compile(optimizer = otimizador, loss = 'binary_crossentropy', metrics = ['binary_accuracy'])
    
    return classificador

# build_fn = funo que cria a rede neural
classificador = KerasClassifier(build_fn = criarRede, epochs = 100, batch_size = 10)

# Corss-validation
# cv = quantidade de vezes que realiza a diviso de treino e teste e executa o treinamento
# scoring = para cada treinamento guarda o scoring obtido, no caso a acurcia (avaliao dos resultados)
resultados = cross_val_score(estimator = classificador, X = previsores, y = classe, cv = 10, scoring = 'accuracy')

# Mdia e desvio padro do scoring (acurcia) de todos os 10 resultados
media = resultados.mean()
desvio = resultados.std()
```
