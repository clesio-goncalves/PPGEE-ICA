```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
```

```python
base = pd.read_csv('iris.csv')
previsores = base.iloc[:, 0:4].values
classe = base.iloc[:, 4].values
```

```python
# Transformação entre atributos categóricos e atributos numéricos (0, 1 e 2)
labelencoder = LabelEncoder()
classe = labelencoder.fit_transform(classe)

# Transformação de dados numéricos para dados binários categóricos (100, 010, 001), pois a rede possui 3 neurônios na camada de saída
classe_dummy = np_utils.to_categorical(classe)
```

```python
def criar_rede():

    # Cria a rede neural
    classificador = Sequential()
    classificador.add(Dense(units = 4, activation = 'relu', input_dim = 4))
    classificador.add(Dense(units = 4, activation = 'relu'))
    classificador.add(Dense(units = 3, activation = 'softmax'))
    classificador.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
    return classificador

# Cria o classificador com 1000 epochs e 10 batch_size
classificador = KerasClassifier(build_fn = criar_rede, epochs = 1000, batch_size = 10)
```

```python
# Cros validation com 10 divisões na base de dados (cv = 10)
# O scoring é avaliado utilizando a métrica acurácia
# Gera um vetor com 10 resultados de acurácia (1 resultado para cada divisão em treino e teste)
resultados = cross_val_score(estimator = classificador, X = previsores, y = classe, cv = 10, scoring = 'accuracy')
```

```python
# Média dos 10 resultados do vetor de acurácia
media = resultados.mean()
```

```python
# Desvio padrão dos 10 resultados do vetor de acurácia
# Verifica o quanto os dados estão variando em relação a média para analisar se há overfitting na base de dados
desvio = resultados.std()
```
