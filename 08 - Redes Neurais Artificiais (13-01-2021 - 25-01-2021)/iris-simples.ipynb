```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
# Leitura da base e divisão entre treino e teste
base = pd.read_csv('iris.csv')
previsores = base.iloc[:, 0:4].values
classe = base.iloc[:, 4].values
```

```python
# É necessário realizar a transformação dos dados String em dados numéricos (0, 1 e 2), pois a rede neural trabalha com dados numéricos
labelencoder = LabelEncoder()
classe = labelencoder.fit_transform(classe)

# É necessário transformar os dados inteiros com 1 classe de saída [shape (1,)] em dados binários categóricos com 3 classes de saída [shape (3,)]
# Transformação para dados categóricos binários --> ao aplicar problemas de classificação com mais de 2 classes de saída
classe_dummy = np_utils.to_categorical(classe)
# iris setosa     	1 0 0
# iris virginica  	0 1 0
# iris versicolor 	0 0 1
```

```python
# Divisão da base de dados em treino e teste
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classe_dummy, test_size=0.25)
```

```python
# Rede neural
classificador = Sequential()

# (entrada + saida) / 2  --> (4 + 3) / 2 = 4 neurônios na camada oculta
# primeira camada oculta, conectada a 4 neurônios na camada de entrada
classificador.add(Dense(units = 4, activation = 'relu', input_dim = 4)) 

# segunda camada oculta com 4 neurônios
classificador.add(Dense(units = 4, activation = 'relu'))

# camada de saída (3 neurônios --> 3 classes de saída)
# função de ativação sigmoid --> problemas com 2 classes de saída, pois retorna uma probabilidade entre 0 e 1
# função de ativação softmax --> utilizada em problemas de classificação com mais de 2 classes de saída
# softmax --> gera uma probabilidade para cada uma das classes de saída, classificando a classe de maior valor de probabilidade
classificador.add(Dense(units = 3, activation = 'softmax'))


# adam --> melhoria na descida do gradiente estocástica
# categorical_crossentropy --> função de perda para problemas de classificação com mais de 2 classes de saída (pode usar a função kullback)
# categorical_accuracy --> acurácia para classificação com mais de 2 classes de saída
classificador.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])

# Treinamento da rede neural
# epochs = 1000 --> mil atualizações de pesos
# durante o treinamento da rede, é exibido o valor da acurácia. Mas esse valor corresponde a acurácia na base de treinamento. 
classificador.fit(previsores_treinamento, classe_treinamento, batch_size = 10, epochs = 1000)
```

```python
# Avaliação dos dados de teste (retorna a função de perda e acurácia) --> função do keras
resultado = classificador.evaluate(previsores_teste, classe_teste)
```

```python
# Predição com os dados de teste --> função manual
previsoes = classificador.predict(previsores_teste)

# Limiar previsões que retorna true ou false (0 ou 1)
previsoes = (previsoes > 0.5)
```

```python
# Retorna o indice que possui o valor 1, ou seja, o resultado da classificação
classe_teste2 = [np.argmax(t) for t in classe_teste]
previsoes2 = [np.argmax(t) for t in previsoes]

# Matriz de confusão das 3 classes
matriz = confusion_matrix(previsoes2, classe_teste2)
```
