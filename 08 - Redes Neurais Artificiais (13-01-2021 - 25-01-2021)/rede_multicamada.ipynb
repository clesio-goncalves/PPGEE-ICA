```python
import numpy as np
```

```python
epocas = 100000  # Tempo de treinamento
momento = 1
taxa_aprendizagem = 0.7
```

```python
def sigmoid(soma):
    return 1 / (1 + np.exp(-soma)) # Numero de euler (e)
```

```python
def sigmoidDerivada(sig):
    return sig * (1 - sig)
```

```python
# XOR  --->  Não linearmente separável
# XOR necessita de camada(s) oculta(s)
# qnt_neuronios_camada_oculta = (entradas + saidas) / 2
# 2 camadas ocultas funcionam bem pra poucos dados
entradas = np.array([[0,0],
                     [0,1],
                     [1,0],
                     [1,1]])

saidas = np.array([[0],[1],[1],[0]])
```

```python
# 2 neurônios na camada de entrada, 3 neuronios na camada oculta e 1 neurônio na camada de saída

# Pesos para X1 e X2 (da camada de entrada para a camada oculta com 3 neurônios)
pesos0 = 2 * np.random.random((2, 3)) - 1 # gera valores aleatorios positivos e negativos

pesos1 = 2 * np.random.random((3, 3)) - 1 # gera valores aleatorios positivos e negativos

# Pesos da camada oculta para a camada de saída
pesos2 = 2 * np.random.random((3, 1)) - 1 # gera valores aleatorios positivos e negativos
```

```python
for j in range(epocas):
    
    # Sinapse da camada de entrada para a camada oculta 1
    camadaEntrada = entradas
    somaSinapse0 = np.dot(camadaEntrada, pesos0)
    camadaOculta1 = sigmoid(somaSinapse0)
    
    # Sinapse da camada oculta 1 para a camada oculta 2
    somaSinapse1 = np.dot(camadaOculta1, pesos1)
    camadaOculta2 = sigmoid(somaSinapse1)
    
    # Sinapse da camada oculta para a camada de saída
    somaSinapse2 = np.dot(camadaOculta2, pesos2)
    camadaSaida = sigmoid(somaSinapse2)
    
    # Erro da camada de saída
    erroCamadaSaida = saidas - camadaSaida
    MSE = np.mean(np.power(erroCamadaSaida, 2))
    print("Erro: " + str(MSE))
    
    
    # ---------------------------------- Backpropagation ----------------------------------------- #
    
    
    # Delta da camada de saída
    # Erro camada de saída * derivada sigmoide da camada de saída
    derivadaSaida = sigmoidDerivada(camadaSaida)
    deltaSaida = erroCamadaSaida * derivadaSaida
    
    # Delta da camada oculta 2
    # Delta da camada de saída * pesos da camada oculta 2 * derivada sigmoide da camada oculta 2
    pesos2Transposta = pesos2.T
    deltaSaidaXPeso2 = deltaSaida.dot(pesos2Transposta)
    deltaCamadaOculta2 = deltaSaidaXPeso2 * sigmoidDerivada(camadaOculta2)
    
    # Delta da camada oculta 1
    # Delta da camada oculta 2 * pesos da camada oculta 1 * derivada sigmoide da camada oculta 1
    pesos1Transposta = pesos1.T
    deltaCamadaOculta2XPeso1 = deltaCamadaOculta2.dot(pesos1Transposta)
    deltaCamadaOculta1 = deltaCamadaOculta2XPeso1 * sigmoidDerivada(camadaOculta1)
    
    # Atualização dos pesos da camada de saída para a camada oculta 2
    camadaOculta2Transposta = camadaOculta2.T
    pesosNovo2 = camadaOculta2Transposta.dot(deltaSaida) # somatorio (camada oculta 2 * delta saída)
    pesos2 = (pesos2 * momento) + (pesosNovo2 * taxa_aprendizagem) # atualização dos pesos
    
    # Atualização dos pesos da camada oculta 2 para a camada oculta 1
    camadaOculta1Transposta = camadaOculta1.T
    pesosNovo1 = camadaOculta1Transposta.dot(deltaCamadaOculta2) # somatorio (camada oculta 1 * delta camada oculta 2)
    pesos1 = (pesos1 * momento) + (pesosNovo1 * taxa_aprendizagem) # atualização dos pesos
    
    # Atualização dos pesos da camada oculta 1 para a camada de entrada
    camadaEntradaTransposta = camadaEntrada.T
    pesosNovo0 = camadaEntradaTransposta.dot(deltaCamadaOculta1) # somatorio (camada de entrada * delta camada oculta 1)
    pesos0 = (pesos0 * momento) + (pesosNovo0 * taxa_aprendizagem) # atualização dos pesos
    
```
